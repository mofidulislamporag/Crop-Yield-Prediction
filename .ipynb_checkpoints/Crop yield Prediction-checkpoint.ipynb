```python
# Crop Yield Prediction System
# Comprehensive analysis and machine learning model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('default')
sns.set_palette("husl")

print("âœ… Crop Yield Prediction System Initialized")
```

```python
# Step 1: Load and Explore the Data
print("ðŸ“Š Loading and Exploring Dataset...")

# Load the dataset
df = pd.read_csv('crop_yield.csv')

# Basic information about the dataset
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

print("\nDataset Info:")
df.info()

print("\nBasic Statistics:")
display(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Step 2: Data Visualization and Analysis
print("ðŸ“ˆ Data Visualization and Analysis...")

# Create subplots for better visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 1. Yield distribution
axes[0,0].hist(df['Yield_tons_per_hectare'], bins=30, alpha=0.7, color='skyblue')
axes[0,0].set_title('Yield Distribution')
axes[0,0].set_xlabel('Yield (tons/hectare)')
axes[0,0].set_ylabel('Frequency')

# 2. Crop-wise yield
crop_yield = df.groupby('Crop')['Yield_tons_per_hectare'].mean().sort_values(ascending=False)
crop_yield.plot(kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('Average Yield by Crop Type')
axes[0,1].set_ylabel('Average Yield (tons/hectare)')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Region-wise yield
region_yield = df.groupby('Region')['Yield_tons_per_hectare'].mean().sort_values(ascending=False)
region_yield.plot(kind='bar', ax=axes[0,2], color='lightcoral')
axes[0,2].set_title('Average Yield by Region')
axes[0,2].set_ylabel('Average Yield (tons/hectare)')
axes[0,2].tick_params(axis='x', rotation=45)

# 4. Soil type yield
soil_yield = df.groupby('Soil_Type')['Yield_tons_per_hectare'].mean().sort_values(ascending=False)
soil_yield.plot(kind='bar', ax=axes[1,0], color='gold')
axes[1,0].set_title('Average Yield by Soil Type')
axes[1,0].set_ylabel('Average Yield (tons/hectare)')
axes[1,0].tick_params(axis='x', rotation=45)

# 5. Weather condition yield
weather_yield = df.groupby('Weather_Condition')['Yield_tons_per_hectare'].mean().sort_values(ascending=False)
weather_yield.plot(kind='bar', ax=axes[1,1], color='violet')
axes[1,1].set_title('Average Yield by Weather Condition')
axes[1,1].set_ylabel('Average Yield (tons/hectare)')
axes[1,1].tick_params(axis='x', rotation=45)

# 6. Fertilizer usage impact
fert_yield = df.groupby('Fertilizer_Used')['Yield_tons_per_hectare'].mean()
fert_yield.plot(kind='bar', ax=axes[1,2], color='orange')
axes[1,2].set_title('Yield: Fertilizer Used vs Not Used')
axes[1,2].set_ylabel('Average Yield (tons/hectare)')
axes[1,2].set_xticklabels(['No Fertilizer', 'Fertilizer'], rotation=0)

plt.tight_layout()
plt.show()
```
